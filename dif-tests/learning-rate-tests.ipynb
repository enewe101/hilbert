```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_losses(fname):
    losses = []
    with open(fname, 'rt', encoding='utf-8') as f:
        for line in f:
            if line[:4] == 'iter':
                losses.append(float(line.split('=')[1]))
    return losses
```

```python
def plot_loss(fname):
    losses = load_losses(fname)
    plt.plot(range(len(losses)), losses, label=fname)
    return losses
    
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 22})
# plot_loss('/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/1.8-5w-mle-l500-I200-V10k/trace.txt')
# plot_loss('/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/1.32-5w-mle-l500-I200-V10k/trace.txt')
# plot_loss('/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/1.128-5w-mle-l500-I300-V10k/trace.txt')
# plot_loss('/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/1.512-5w-mle-l500-I300-V10k/trace.txt')
plot_loss('/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/1.8-5w-map-l500-I200-V10k/trace.txt')
#plt.xlim(10000, 30000)
# plt.ylim(,4)
plt.legend()
plt.show()
```
