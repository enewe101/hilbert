```python
import numpy as np
import matplotlib.pyplot as plt
from os import listdir
```

```python
def load_losses(fname):
    losses = []
    with open(fname, 'rt', encoding='utf-8') as f:
        for line in f:
            if line[:4] == 'iter':
                losses.append(float(line.split('=')[1]))
    return losses
```

```python
def plot_loss(fname, norm=True):
    path_temp = '/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/{0}/trace.txt'
    full_path = path_temp.format(fname)
    losses = load_losses(full_path)
    
    x_coords = range(len(losses))
    if norm:
        x_coords = np.array(x_coords) / x_coords[-1]
        
    plt.plot(x_coords, losses, label=fname)
    return losses
    
```

```python
def find_fnames(match):
    dirs = listdir('/home/rldata/hilbert-embeddings/embeddings/diffusion-hilbert/')
    matched_dirs= [ d for d in dirs if match in d]
    return matched_dirs
```

```python
plt.figure(figsize=(20,10))
plt.rcParams.update({'font.size': 22})

dirs = find_fnames('9w-mle')
for d in dirs:
    plot_loss(d,False)

# plot_loss('1.2048-1w-diff1-l50-I400-V10k', False)
# plot_loss('1.2048-1w-diff1-l100-I400-V10k', False)
# plt.ylim(,4)
plt.legend(loc='upper left')
plt.show()
```
