```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_losses(fname):
    losses = []
    with open(fname, 'rt', encoding='utf-8') as f:
        for line in f:
            if line[:4] == 'iter':
                losses.append(float(line.split('=')[1]))
    return losses
```

```python
def plot_loss(fname):
    losses = load_losses(fname)
    plt.plot(range(len(losses)), losses, label=fname)
    return losses
    
```
